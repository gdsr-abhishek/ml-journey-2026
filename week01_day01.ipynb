```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def add_vectors(v1,v2):
    return [v1[i]+v2[i] for i in range(len(v1))]
def numpy_add_vectors(v1,v2):
    return np.array(v1) + np.array(v2)
```

```python
print(add_vectors([1,2],[2,3]))
```

```python
origin = [0,0]
v1,v2=[2,1],[1,3]
result  = add_vectors(v1,v2)

#plt
plt.quiver(*origin,*v1,color='blue',scale=1,scale_units='xy',angles='xy')
plt.quiver(*v1,*v2,color='green',scale=1,scale_units='xy',angles='xy')
plt.quiver(*origin,*result,color='red',scale=1,scale_units='xy',angles='xy')
plt.xlim(0,5);plt.ylim(0,5);plt.grid()
plt.show()
```

```python
def dot_prod(v1,v2):
    return sum([v1[i]*v2[i] for i in range(len(v1))])
```

```python
v3=[2,5]
v4=[2,-1]
dot =dot_prod(v3,v4)
np_dot = np.dot(np.array(v3),np.array(v4))
print(f"np dot:{np_dot} custom dot product: {dot}")
plt.quiver(*origin,*v3,color='blue',scale=1,scale_units='xy',angles='xy')
plt.quiver(*origin,*v4,color='green',scale=1,scale_units='xy',angles='xy')
plt.xlim(-5,5);plt.ylim(-5,5);plt.grid()
plt.show()
```

```python
def rotation(v, theta_degree):
    theta = np.radians(theta_degree)
    R = np.array([
        [np.cos(theta), -np.sin(theta)],
        [np.sin(theta), np.cos(theta)]
    ])
    return R @ np.array(v)
```

```python
rotation([0,1],90)
```

```python
# create a grid of points
x = np.linspace(-1, 1, 5)
y = np.linspace(-1, 1, 5)
points = np.array([[xi, yi] for xi in x for yi in y]).T  # 2×N

# rotate all of them
theta = np.radians(45)
R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])

rotated = R @ points

# plot original vs rotated
plt.scatter(points[0], points[1], color='blue', label='original')
plt.scatter(rotated[0], rotated[1], color='red', label='rotated 45°')
plt.legend(); plt.grid(); plt.axis('equal')
plt.show()
```
